```python
import json, os
from datetime import datetime

DB_FILE = "db.json"

def init_db():
    if not os.path.exists(DB_FILE):
        data = {
            "patients": [],
            "doctors": [
                {"id": 1, "name": "Dr. A. Kumar", "specialization": "General"},
                {"id": 2, "name": "Dr. S. Rao", "specialization": "Pediatrics"}
            ],
            "appointments": [],
            "bills": []
        }
        with open(DB_FILE, "w") as f:
            json.dump(data, f, indent=4)

def read_db():
    return json.load(open(DB_FILE))

def save_db(data):
    json.dump(data, open(DB_FILE, "w"), indent=4)

init_db()

# Patient
def add_patient(name, phone, age, gender):
    db = read_db()
    pid = len(db["patients"]) + 1
    db["patients"].append({"id": pid, "name": name, "phone": phone, "age": age, "gender": gender})
    save_db(db)
    return pid

def get_patients():
    return read_db()["patients"]

# Doctors
def get_doctors():
    return read_db()["doctors"]

# Appointments
def add_appointment(pid, did, dt, reason):
    db = read_db()
    aid = len(db["appointments"]) + 1
    db["appointments"].append({"id": aid, "patient_id": pid, "doctor_id": did, "datetime": dt, "reason": reason})
    save_db(db)
    return aid

def get_appointments():
    return read_db()["appointments"]

# Bills
def add_bill(pid, amount, details):
    db = read_db()
    bid = len(db["bills"]) + 1
    db["bills"].append({"id": bid, "patient_id": pid, "amount": amount, "details": details, "created_at": str(datetime.now())})
    save_db(db)
    return bid

def get_bills():
    return read_db()["bills"]
```
